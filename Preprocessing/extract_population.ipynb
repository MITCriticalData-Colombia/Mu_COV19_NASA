```python
# Import necessary modules
import geopandas as gpd
import pandas as pd
```

```python
# Add path to folder with shapefile
path = '../Data/Census_data/'
out_path = '../Data/Census_data.csv'
```

```python
# Read file using gpd.read_file()
data = gpd.read_file(path)
data.head(3)
```

```python
for i, datatype in enumerate(data.dtypes):
    print(f'column: {data.columns[i]} has datatype: {datatype}')
```

```python
columns = ["total_population", "Municipality_code", "latitude", "longitude", 'geometry']
data = data.rename(columns={"STP27_PERS":"total_population", "DPTO_CNMBR": "Municipality_code", "LATITUD": "latitude", "LONGITUD": "longitude"})

data = data[columns]
```

```python
data.rename(columns={'Municipality_code':'Dpto'}, inplace=True)
data
```

```python
data.Dpto = data.Dpto.str.title()

data.Dpto = data.Dpto.str.replace(" ", "")

a,b = 'áéíóúüñÁÉÍÓÚÜÑ','aeiouunAEIOUUN'
trans = str.maketrans(a,b)
data.Dpto = data.Dpto.str.translate(trans)

data.Dpto[data.Dpto == 'Bogota,D.C.'] = "Bogota"
data.Dpto[data.Dpto == 'ArchipielagoDeSanAndres,ProvidenciaYSantaCatalina'] = 'SanAndres'
data.Dpto = data.Dpto.str.replace("D", "d")

data.Dpto = data.Dpto.str.lower()
data.Dpto = data.Dpto.str.strip()
```
