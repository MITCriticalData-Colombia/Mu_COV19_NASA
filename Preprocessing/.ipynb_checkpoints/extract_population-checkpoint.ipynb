# <a href="https://geoportal.dane.gov.co/servicios/descarga-y-metadatos/visor-descarga-geovisores/"> Population Dataset shapefile format<a/> 

```python
# Import necessary modules
import geopandas as gpd
```

```python
# Add path to folder with shapefile
path = '../Data/Census_data/'
out_path = '../Data/Census_data.csv'
```

```python
# Read file using gpd.read_file()
data = gpd.read_file(path)
data.head(3)
```

```python
for i, datatype in enumerate(data.dtypes):
    print(f'column: {data.columns[i]} has datatype: {datatype}')
```

```python
columns = ["total_population", "Municipality_code", "latitude", "longitude"]
data = data.rename(columns={"STP27_PERS":"total_population", "MPIO_CDPMP": "Municipality_code", "LATITUD": "latitude", "LONGITUD": "longitude"})
data = data[columns]
```

```python
# Download Data as csv
data.to_csv(out_path)
```
